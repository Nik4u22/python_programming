```python
s = 'this is the python class'
s1 = 'this is THE python class'

# capitalize - it capitalises the first letter of the sentence and lower case all the other char
print(s.capitalize())
print(s1.capitalize())

#upper - converts to upper case
print(s.upper())
print(s1.upper())

#lower - converts to lower case
print(s.lower())
print(s1.lower())

# Swapcase - it converts uppercase to lopwerase and lowercase to uppercase
print(s.swapcase())
print(s1.swapcase())

#title - converts each words first letter to uppercase
print(s.title())
print(s1.title())
```

```python
# functions that return a bool value

# 1. startwith --> Return True if string starts with the specified prefix else False

s = 'hello world'
print(s.startswith('ello'))

print(s.startswith('o ',4,10))

# 2. ends with - return True if string ends with a specified suffix else false
s = 'hello world'
print(s.endswith('rld'))

print(s.endswith('rl',4,10))
```

```python
# strip, rstrip , lstrip

# strip - removes the specified char from both the ends
a = '   pyhton  '
print(a)
print(a.strip()) #deafult is space

a = '***pyhton*****'
print(a)
print(a.strip('*')) #deafult is space
```

```python
# rstrip - removes the specified char from the right side
a = '***pyhton*****'
print(a)
print(a.rstrip('*')) #deafult is space
```

```python
# lstrip - removes the specified char from the left side
a = '***pyhton*****'
print(a)
print(a.lstrip('*')) #deafult is space
```
