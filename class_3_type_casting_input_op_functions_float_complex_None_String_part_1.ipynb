# Complex data type

- any value that can be written in the form of a+bj where a is the real value and b is the imaginary value

```python
a = 10+7j

print(type(a))
```

```python
# real
print(a.real)
```

```python
# imag
print(a.imag)
```

```python
# obj reusabilty is not possible

a = 10+6j
b = 10+6j

print(a is b)
```

```python
# Operation are possible with Complex

- arithamatic
- logical
- Equality
- indentity
- assignment op

# opertation that are not possible with float

- bitwise
- ascii
-comparison
```

```python
a = 10+6j
b = 10+6j

print(a > b)
```
