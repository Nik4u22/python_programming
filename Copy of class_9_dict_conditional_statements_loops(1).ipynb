1. get

```python
# get

if the key is preset it returns the value, else deafult value be returend
```

```python
# a.) key is present

a = {'emp1':{'firstname':'afsan', 'age':25 , 'salary':25},
    'emp2':{'firstname':'rahul', 'age':25 , 'salary':60}}

a.get('emp1')
```

```python
# a.) key is not present

a = {'emp1':{'firstname':'afsan', 'age':25 , 'salary':25},
    'emp2':{'firstname':'rahul', 'age':25 , 'salary':60}}

print(a.get('emp3'))
```

```python
# customise the message
a = {'emp1':{'firstname':'afsan', 'age':25 , 'salary':25},
    'emp2':{'firstname':'rahul', 'age':25 , 'salary':60}}

print(a.get('emp3' , 'key not present'))
```

```python
a.get('emp1')
```

```python
a.get('emp1').get('name')
```

```python
a.get('emp1').get('age')
```

```python
2. setdeafult

insert the key with the deafult value if the key is not present, else return the value of the key
```

```python
# a.) key is not present

a = {'emp1':{'firstname':'afsan', 'age':25 , 'salary':25},
    'emp2':{'firstname':'rahul', 'age':25 , 'salary':60}}

print(a.setdefault('emp3' , 'key not present'))
```

```python
a
```

```python
# a.) key is present

a = {'emp1':{'firstname':'afsan', 'age':25 , 'salary':25},
    'emp2':{'firstname':'rahul', 'age':25 , 'salary':60}}

print(a.setdefault('emp1' , 'key not present'))
```

```python
a = {'emp1':{'firstname':'afsan', 'age':25 , 'salary':25},
    'emp2':{'firstname':'rahul', 'age':25 , 'salary':60}}
```

```python
a.get('emp1').get('firstname') = 'afsan'
```

```python
a['emp2']['lastname'] = 'a'
```

```python
a
```
