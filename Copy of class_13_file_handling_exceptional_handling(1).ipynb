## 1. Read mode - 'r'
- If the file exists then it will open the file in read mode and places the cursor at the begining of the file
- if the file does not exists then it will give FileNotFoundError
- You are allowed only to read from the file but not allowed to write into the file

```python
# file is not present

f = open('read.py','r')
f.tell()
```

```python
# file is present

f = open('read.py','r')
f.tell()
f.read()
f.tell()
```

```python
"""
this is for testing read mode
"""

f = open('read.py','r')
print(f.tell()) # 0
print(f.read()) # this is for testing read mode
print(f.seek(10)) #
print(f.read()) # r testing read mode
print(f.tell()) # eof
f.write('abc') #
```

## 2. Write mode - 'w'
- If the file exists then it will open the file in write mode and deletes the content of the file
- if the file does not exists then it will create a file with given name and extension
- You are allowed only to write into the file but not allowed to read from the file

```python
# if file is not preset

f = open('write.py','w')
```

```python
# if the file is present
f = open('write.py','w')
f.tell() # 0
f.write('this is for testing purpose')
f.tell()
```

```python
# if the file is present
"""
hi
"""
f = open('write.py','w')
f.write('hi')
f.tell() # eof
f.seek(1)
f.tell() # 1
f.write('hey wassup')
f.close()
f.read()
```

## 3. Append mode - 'a'
- If the file exists then it will open the file in write mode and places the cursor at the end of the file
- if the file does not exists then it will create a file with given name and extension
- You are allowed only to write into the file but not allowed to read from the file
- append will always write the the content at the end of the file so seek() will not work here

```python
# file is not present

f = open('append.py','a')
```

```python
# file is present
"""
for testing purposeappend modemode
"""
f = open('append.py','a')
f.tell() #eof
f.write('append mode')
f.seek(3)
f.tell() # 3
f.write('mode')
f.close()
f.read() # error
```

```python
## 4. Write and read mode - 'w+'
- If the file exists then it will open the file in write mode and deletes the content of the file
- if the file does not exists then it will create a file with given name and extension
- You are allowed to write into the file and also allowed to read from the file
```

```python
# if file is not preset

f = open('write_read.py','w+')
```

```python
# if the file is present
"""
hhey wassupafsan

"""
f = open('write_read.py','w+')
f.write('hi')
print(f.tell()) # eof
f.seek(1)
print(f.tell()) # 1
f.write('hey wassup')
f.seek(2)
print(f.read())
print(f.tell())
f.write('afsan\n')
print(f.read())
f.seek(0)
print(f.read())
f.close()
```

```python
## 5. read and write - 'r+'
- If the file exists then it will open the file in read mode and places the cursor at the begining of the file
- if the file does not exists then it will give FileNotFoundError
- You are allowed to write into the file and also allowed to read from the file
```

```python
# file is not present

f = open('read_write.py','r+')
```

```python
"""
hiey wassupafsanhello
"""
f = open('read_write.py','r+')
print(f.read())
print(f.tell())
f.write('hello')
f.seek(0)
f.write('hi')
print(f.read())
f.write('bye')
print(f.read())
f.close()
```

```python
# 6. Append and read mode - 'a+'

- If the file exists then it will open the file in append mode and places the cursor at the end of the file
- if the file does not exists then it will create a file with given name and extension
- You are allowed to write into the file and also allowed to read from the file
- append will always write the the content at the end of the file so seek() will not here
```
