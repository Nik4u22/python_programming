shallow and deep copy

```python
a = [1,2,3,4,5]
b = [1,2,3,4,5]

print(id(a))
print(id(b))
```

```python
a = [1,2,3,4,5]
b = a

print(id(a))
print(id(b))

b[0] = 100
print(a)
print(b)
```

```python
# shallow copy

a = [1,2,3,4,5]
b = a.copy()

print(a , id(a))
print(b, id(b))

b[0] = 100

print(a , id(a))
print(b, id(b))
```

```python
# problem with the shallow copy is it does not do the same thing when its nested list

a = [1,2,3,[4,5]]
b = a.copy()

print(a , id(a))
print(b, id(b))

b[-1][0] = 100

print(a , id(a))
print(b, id(b))
```

```python
# deep copy
from copy import deepcopy

a = [1,2,3,[4,5]]
b = deepcopy(a)

print(a , id(a))
print(b, id(b))

b[-1][0] = 100

print(a , id(a))
print(b, id(b))
```

```python
a = [1,2,3,[4,5]]
b = a.copy()

print(a , id(a[-1]))
print(b, id(b[-1]))

b[-1][0] = 100

print(a , id(a))
print(b, id(b))
```

```python
a = [1,2,3,[4,5]]
b = deepcopy(a)

print(a , id(a[-1]))
print(b, id(b[-1]))

b[-1][0] = 100

print(a , id(a))
print(b, id(b))
```
