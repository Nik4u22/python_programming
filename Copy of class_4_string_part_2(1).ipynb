```python
# String Properties

1. String is a seq data type - it allows +ve and -ve indexing
2. +ve indexing starts from 0 in l to R direction
3. -ve indexing starts from -1 in R ro L direction
4. Indexing , Slicing , concatination . reptition and membership, identity
5. String is Immutable
```

```python
# indexing

- access one elemet at time
- +ve indexing starts from 0 in l to R direction
- -ve indexing starts from -1 in R ro L direction
```

```python
a = 'python'
print(a[0])
print(a[-1])
```

```python
# slicing - extract the sub seq from a seq

syntax = var[start_index : stop_index : step]
```

```python
a = 'python class'

# +ve
print(a[7:])

# -ve
print(a[-5:])
```

```python
# Default values

when step is +ve
default start index is 0
default stop index is len(str)

when step is -ve
default start index is -1
default stop index is -len(str)-1

step size can be +/-n so this means every nth value
```

```python
# step

when step is +ve --> slicing will be done from left to right
when step is -ve --> slicing will be done from right to left

so step can be

+1 --> every char from left to right
-1 --> every char from right to left

+2 --> every 2nd char from left to right
-2 --> every 2nd char from right to left


+n --> every nth char from left to right
-n --> every nth char from right to left
```
