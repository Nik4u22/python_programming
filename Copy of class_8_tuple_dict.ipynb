```python
# properties of dict

1. coleection of key value pair
2. key,value pair is called as item
3. items are sep by ,
4. key and value are sep :
5. key cannot be duplicated but value can be duplicated
6. key cannot be mutable
7. value can by any data type
8. all elem are encloed inside a {}
9. dict a not a seq dta type - indexing , slicing , concatination , repetition
10. membership op is applicable but only on keys
11. identiy op is possible
12. dic is mutable
```

```python
"""1. coleection of key value pair
2. key,value pair is called as item
3. items are sep by ,
4. key and value are sep :
8. all elem are encloed inside a {}"""

data = {'name': 'afsan', 'age': '25', 'salary': '80'}
print(data , type(data))
```

```python
# 5. key cannot be duplicated but value can be duplicated

# a.) when value is duplicated

data = {'name': 'afsan', 'age': '25', 'salary': '25'}
print(data)
```

```python
# a.) when key is duplicated

data = {'name': 'afsan', 'age': '25', 'age': '30'}
print(data)
```

```python
# 6. key cannot be mutable

data = {'name': 'afsan', 'age': '25', 'age': '30'}
data = {'name': 'afsan', 'age': '25', (10,20,30): '30'}
data = {'name': 'afsan', 'age': '25', 1: '30'}
data = {'name': 'afsan', [1,2,3]: '25', (10,20,30): '30'}
```

```python
# 7. value can by any data type
data = {'name': 'afsan', 'age': 25, (10,20,30): [1,2,3]}
```

```python
# 9. dict a not a seq dta type - indexing , slicing , concatination , repetition

data[0]
```

```python
# 10. membership op is applicable but only on keys
data = {'name': 'afsan', 'age': 25, (10,20,30): [1,2,3]}
print('afsan' in data)
```

```python
# 10. membership op is applicable but only on keys
data = {'name': 'afsan', 'age': 25, (10,20,30): [1,2,3]}
print('age' in data)
```

```python
# 11. identiy op is possible

data = {'name': 'afsan', 'age': 25, (10,20,30): [1,2,3]}
data1 = {'name': 'afsan', 'age': 25, (10,20,30): [1,2,3]}

print(data is data1)
```

```python
#12 . mutable

data = {'name': 'afsan', 'age': 25, (10,20,30): [1,2,3]}
data['name'] = 'rahul'

data
```
