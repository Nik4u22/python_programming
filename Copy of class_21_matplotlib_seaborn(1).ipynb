```python
week_day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,49,46]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(week_day, temp)
```

```python
plt.plot(week_day, temp)
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('day vs temp')
```

```python
plt.plot(week_day, temp, color='black')
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('day vs temp')
```

```python
plt.plot(week_day, temp, color='black', linestyle='--')
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('day vs temp')
```

```python
plt.plot(week_day, temp, color='black', linestyle='--', marker='o')
plt.xlabel('Day')
plt.ylabel('temp')
plt.title('day vs temp')
```

```python
days=[1,2,3,4,5,6,7] # weekdays
max_t=[50,51,52,48,47,49,46] # max temp recorder at that day
min_t=[43,42,40,44,33,35,37] # min temp
avg_t=[45,48,48,46,40,42,41] # avg temp
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend()
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend()
plt.grid()
```

```python
import seaborn as sns
```

```python
dataset = sns.load_dataset('tips')
```

```python
dataset
```

```python
sns.relplot(x = 'total_bill', y='tip', data = dataset)
```

```python
sns.relplot(x = 'total_bill', y='tip', data = dataset, hue = 'smoker') # categrical-->based on color
```

```python
numerical - 1,1,6

categorical- poor , aveerage, ex
group - 2 , 3 ,4
```

```python
sns.relplot(x = 'total_bill', y='tip', data = dataset, hue = 'smoker', style = 'time') 
```

```python
sns.relplot(x = 'total_bill', y='tip', data = dataset, hue = 'smoker', style = 'time', size='sex') 
```

```python
sns.relplot(x = 'total_bill', y='tip', data = dataset, hue = 'smoker', style = 'time', size='sex', col='day')
```

```python
sns.scatterplot(x = 'total_bill', y='tip', data = dataset)
```

```python
sns.barplot(x = 'day', y='tip', data = dataset)
```

```python
sns.catplot(x = 'day', y='tip', data = dataset)
```

```python
sns.catplot(x ='day', y='tip', data = dataset, kind='violin')
```

```python
sns.catplot(x ='day', y='tip', data = dataset, kind='swarm')
```
