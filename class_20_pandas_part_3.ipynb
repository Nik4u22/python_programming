# concatination

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
# axis = 0 - rowise
# axis = 1 - columnwise

new_data = pd.concat((india_weather, us_weather), axis=0)
```

```python
new_data.reset_index(drop=True)
```

```python
pd.concat((india_weather, us_weather), axis=0, ignore_index=True)
```

```python
pd.concat((india_weather, us_weather), axis=1)
```

```python
# 2 level indexing
data = pd.concat((india_weather, us_weather), axis=0,keys=['india','us'])
```

```python
# MERGING

- INNER MERGE
- OUTER MERGE
- LEFT MERGE
- RIGHT MERGE
```

reference:- https://www.w3schools.com/sql/sql_join.asp

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1,df2,how='inner',on='city')
```

```python
pd.merge(df1,df2,how='left',on='city')
```

```python
pd.merge(df1,df2,how='right',on='city')
```

```python
pd.merge(df1,df2,how='outer',on='city')
```

```python
pd.merge(df1,df2,how='outer',on='city',indicator=True)
```
