# iteraing over the dataframes

1. itertuples
2. iterrows
3. iteritems

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
```

```python
df
```

```python
# iter tuples
for row in df.itertuples():
    print(row)
```

```python
for row in df.itertuples():
    if row.Country_name == 'France':
        print(row.Organizations_name)
```

```python
  for row in df.iterrows():
      print(row[1],end='\n\n')
```

```python
for row in df.iteritems():
    print(row,end='\n\n')
```
